```python
#Removing HTML tags

import pandas as pd
from bs4 import BeautifulSoup

# Function to remove HTML tags from a given text
def remove_html_tags(text):
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text()

# Apply the function to the 'Text' column of the dataframe

data['E1 - Examples of Time Series Analysis'] = data['E1 - Examples of Time Series Analysis'].apply(remove_html_tags)
data['E4 - Panel Data'] = data['E4 - Panel Data'].apply(remove_html_tags)
data['E6 - Decison Trees'] = data['E6 - Decison Trees'].apply(remove_html_tags)
data['E8 - Ensemble Trees Overview'] = data['E8 - Ensemble Trees Overview'].apply(remove_html_tags)
data['E10 - REST-API'] = data['E10 - REST-API'].apply(remove_html_tags)
data['E11 - Natural Language Processing Overview'] = data['E11 - Natural Language Processing Overview'].apply(remove_html_tags)

data.head()
```

```python
#Removing URLs

import pandas as pd
import re

# Function to remove URLs from a given text
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub('', text)


# Apply the function to the 'Text' column of the dataframe

data['E1 - Examples of Time Series Analysis'] = data['E1 - Examples of Time Series Analysis'].apply(remove_urls)
data['E4 - Panel Data'] = data['E4 - Panel Data'].apply(remove_urls)
data['E6 - Decison Trees'] = data['E6 - Decison Trees'].apply(remove_urls)
data['E8 - Ensemble Trees Overview'] = data['E8 - Ensemble Trees Overview'].apply(remove_urls)
data['E10 - REST-API'] = data['E10 - REST-API'].apply(remove_urls)
data['E11 - Natural Language Processing Overview'] = data['E11 - Natural Language Processing Overview'].apply(remove_urls)

data.head()
```

```python
#Removing irrelevant characters

import pandas as pd
import re

# Function to remove irrelevant characters
def remove_irrelevant_chars(text):
    cleaned_text = re.sub(r'[^a-zA-Z0-9]', ' ', text)
    cleaned_text = re.sub(r'[^\w\s]', '', text)
    cleaned_text =cleaned_text.replace('\n', ' ')
    return cleaned_text

# Apply the function to the 'Text' column of the dataframe

data['E1 - Examples of Time Series Analysis'] = data['E1 - Examples of Time Series Analysis'].apply(remove_irrelevant_chars)
data['E4 - Panel Data'] = data['E4 - Panel Data'].apply(remove_irrelevant_chars)
data['E6 - Decison Trees'] = data['E6 - Decison Trees'].apply(remove_irrelevant_chars)
data['E8 - Ensemble Trees Overview'] = data['E8 - Ensemble Trees Overview'].apply(remove_irrelevant_chars)
data['E10 - REST-API'] = data['E10 - REST-API'].apply(remove_irrelevant_chars)
data['E11 - Natural Language Processing Overview'] = data['E11 - Natural Language Processing Overview'].apply(remove_irrelevant_chars)

data.head()
```

```python
#Removing punctuation

import pandas as pd
import string

# Function to remove punctuation
def remove_punctuation(text):
    cleaned_text = text.translate(str.maketrans("", "", string.punctuation))
    return cleaned_text

# Apply the function to the 'Text' column of the dataframe

data['E1 - Examples of Time Series Analysis'] = data['E1 - Examples of Time Series Analysis'].apply(remove_punctuation)
data['E4 - Panel Data'] = data['E4 - Panel Data'].apply(remove_punctuation)
data['E6 - Decison Trees'] = data['E6 - Decison Trees'].apply(remove_punctuation)
data['E8 - Ensemble Trees Overview'] = data['E8 - Ensemble Trees Overview'].apply(remove_punctuation)
data['E10 - REST-API'] = data['E10 - REST-API'].apply(remove_punctuation)
data['E11 - Natural Language Processing Overview'] = data['E11 - Natural Language Processing Overview'].apply(remove_punctuation)

data.head()
```

```python
#Removing duplicates

import pandas as pd

# Remove duplicate text
def remove_duplicate_text(data):
    data = data.drop_duplicates()
    return data

data = remove_duplicate_text(data)

data.head()
```

```python
module_url = "https://tfhub.dev/google/universal-sentence-encoder/2" #@param ["https://tfhub.dev/google/universal-sentence-encoder/2", "https://tfhub.dev/google/universal-sentence-encoder-large/3"]

# Import the Universal Sentence Encoder's TF Hub module
embed = hub.Module(module_url)
```

```python
E1_wsw_df = pd.DataFrame({'E1_wsw': [''.join(df['E1_wsw'].tolist())]})
E4_wsw_df = pd.DataFrame({'E4_wsw': [''.join(df['E4_wsw'].tolist())]})
E6_wsw_df = pd.DataFrame({'E6_wsw': [''.join(df['E6_wsw'].tolist())]})
E8_wsw_df = pd.DataFrame({'E8_wsw': [''.join(df['E8_wsw'].tolist())]})
E10_wsw_df = pd.DataFrame({'E10_wsw': [''.join(df['E10_wsw'].tolist())]})
E11_wsw_df = pd.DataFrame({'E11_wsw': [''.join(df['E11_wsw'].tolist())]})
```

```python
E1_wsw = E1_wsw_df['E1_wsw'].loc[E1_wsw_df.index[0]]
E4_wsw = E4_wsw_df['E4_wsw'].loc[E4_wsw_df.index[0]]
E6_wsw = E6_wsw_df['E6_wsw'].loc[E6_wsw_df.index[0]]
E8_wsw = E8_wsw_df['E8_wsw'].loc[E8_wsw_df.index[0]]
E10_wsw = E10_wsw_df['E10_wsw'].loc[E10_wsw_df.index[0]]
E11_wsw = E11_wsw_df['E11_wsw'].loc[E11_wsw_df.index[0]]
```

```python
with tf.Session() as session:
    session.run([tf.global_variables_initializer(), tf.tables_initializer()])
    sentences_embeddings = session.run(embed([E1_wsw,E4_wsw,E6_wsw,E8_wsw,E10_wsw,E11_wsw]))
```
